```python
from nltk import ngrams
from collections import defaultdict
def count_pair_frq(n_gram):
    pair_freq=defaultdict(int)
    for pair in n_gram:
        pair_freq[pair]+=1
    return pair_freq
sentence=input('enter a sentence')
words=ngrams(sentence.split(),1)
s=count_pair_frq(words)
for pair,freq in s.items():
    print(pair,"->",freq)
```

```python
import nltk
from nltk.probability import FreqDist
text1="hi this is karthik karthik is a good boy karthik is from india india is large country"
tokens=nltk.word_tokenize(text1)
fdist=FreqDist(tokens)
total_count=fdist.N()
freq_count=fdist.B()
text_length=len(text1)
print(tokens)
print('\n')
print(total_count)
print('\n')
print(freq_count)
print('\n')
fdist.plot()
mostcomm=fdist.most_common()
for item,frequency in mostcomm:
    print(item,"->",frequency)
```
